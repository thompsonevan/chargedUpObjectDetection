```python
import cv2
import os
import matplotlib.pyplot as plt
```

```python
file = 'data.mp4'
cap = cv2.VideoCapture(file)

frames = []
count = 0

while(cap.isOpened()): 
    ret, frame = cap.read()
    if(ret):
        frame = cv2.rotate(frame, cv2.ROTATE_180)
        frames.append(frame)
        cv2.imwrite('img' + str(count)+'.jpg', frame)
        count += 1
    else:
        break
```

```python
plt.figure(figsize=(10,10)) # specifying the overall grid size

for i in range(1000):
    plt.subplot(100,100,i+1)    # the number of images in the grid is 5*5 (25)
    plt.imshow(frames[i])

plt.show()
```
